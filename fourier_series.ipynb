```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the square wave
frequency = 5  # Frequency of the square wave
sampling_rate = 1000  # Sampling rate in Hz
t = np.linspace(0, 1, sampling_rate, endpoint=False)  # Time vector

# Generate the square wave
square_wave = np.sign(np.sin(2 * np.pi * frequency * t))

# Plotting the square wave
plt.figure(figsize=(6, 3))
plt.plot(t, square_wave, label='Square Wave')
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the square wave
frequency = 1  # Frequency of the square wave (Hz)
n_terms = 2  # Number of terms in the Fourier series
t = np.linspace(0, 2, 2000, endpoint=False)  # Time vector over 2 seconds at 1000 Hz sampling rate
square_wave = np.zeros_like(t)

# Calculating the square wave using Fourier series
for k in range(1, n_terms + 1):
    harmonic_number = 2 * k - 1  # Only odd harmonics contribute
    square_wave += (4 / (np.pi * harmonic_number)) * np.sin(2 * np.pi * harmonic_number * frequency * t)

# Plotting the square wave
plt.figure(figsize=(6, 3))
plt.plot(t, square_wave)
plt.title('Square Wave Using Fourier Series')
plt.ylim(-1.5, 1.5)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
import ipywidgets as widgets

# Parameters for the square wave
frequency = 1  # Frequency of the square wave (Hz)
t = np.linspace(0, 2, 2000, endpoint=False)  # Time vector over 2 seconds at 1000 Hz sampling rate

def plot_fourier_series(n_terms):
    # Generate the ideal square wave
    ideal_square_wave = np.sign(np.sin(2 * np.pi * frequency * t))
    
    # Fourier series approximation
    square_wave = np.zeros_like(t)
    
    # Summing the Fourier series terms
    for k in range(1, n_terms + 1):
        harmonic_number = 2 * k - 1  # Only odd harmonics contribute
        square_wave += (4 / (np.pi * harmonic_number)) * np.sin(2 * np.pi * harmonic_number * frequency * t)
    
    # Plotting both waves
    plt.figure(figsize=(5, 3))
    plt.plot(t, ideal_square_wave, linestyle='--', color='red')
    plt.plot(t, square_wave)
    plt.title('Fourier Series Approximation')
    plt.ylim(-1.5, 1.5)
    plt.show()

# Interactive widget
interact(plot_fourier_series, n_terms=IntSlider(value=1, min=1, max=200, step=1, description='Number of Terms:'))
```
